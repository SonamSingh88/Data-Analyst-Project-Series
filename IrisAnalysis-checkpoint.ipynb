```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
x,y=iris.data, iris.target
```

```python
print(x[:10])
print(y[:10])
```

```python
df=pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target']=iris.target
print(df.head())
```

```python
df.to_excel('iris_data.xlsx', index=False)
```

```python
df.to_csv('iris_cleaned.csv', index=False)
```

```python
summary=df.describe()
print (summary)
summary_stats="summary.xlsx"
summary.to_excel(summary_stats)
```

```python
print(df.isnull().sum())
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.title("Box Plot of Iris Dataset Features")
plt.show()
```

```python
df.hist(figsize=(12,8))
plt.suptitle("Histograms of Iris Dataset Features")
plt.show()
```

```python
pair_plot=sns.pairplot(df, hue='target')
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

```python
plt.savefig('pair_plot.jpg')
```

```python
grouped_stats=df.groupby('target').agg(['mean', 'median', 'std'])
print(grouped_stats)
grouped="grouped_stats.xlsx"
grouped_stats.to_excel(grouped)
```

```python
for column in df.columns[:-1]:
    sns.displot(df, x=column, hue="target", kind="kde", fill=True)
    plt.title(f'Density Plot for {column}')
    plt.show()
```

```python
correlation_matrix=df.corr()
print(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
CM="correlation_matrix.xlsx"
correlation_matrix.to_excel(CM)
```
